```r
#此例的母體為(4,7,7,8,9)
##demo 中央極限定理
simunum=100000
n=10000
x_bar=1:simunum #到後面每一個都會被取代
for (sj in 1:simunum){
   xdata=sample(c(5,7,8,9),n,prob=c(1/5,2/5,1/5,1/5),replace=T)
   x_bar[sj]=sum(xdata/n) #這樣就會有10000 x_bar, 每一個x_bar是1000個 X_i 的平均

}
hist(   (x_bar-8)/(sqrt(14/4)/sqrt(n))   )
```
